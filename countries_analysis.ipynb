```python
# ==========================================
# Countries Dataset Analysis using Pandas
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np

# Read CSV File
df = pd.read_csv("Countries.csv")

# 1. Display first 5 rows
print("\n========== FIRST 5 ROWS ==========")
print(df.head())
```

```python
# 2. Display last 5 rows
print("\n========== LAST 5 ROWS ==========")
print(df.tail())
```

```python
# 3. Shape of Dataset
print("\n========== SHAPE ==========")
print(df.shape)
```

```python
# 4. Column Names
print("\n========== COLUMN NAMES ==========")
print(df.columns)
```

```python
# 5. Dataset Information
print("\n========== DATASET INFORMATION ==========")
df.info()
```

```python
# 6. Statistical Summary
print("\n========== STATISTICAL SUMMARY ==========")
print(df.describe(include="all"))
```

```python
# 7. Data Types
print("\n========== DATA TYPES ==========")
print(df.dtypes)
```

```python
# 8. Missing Values
print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())
```

```python
# 9. Unique Values
print("\n========== UNIQUE VALUES ==========")
print(df.nunique())
```

```python
# 10. Random 5 Rows
print("\n========== RANDOM 5 ROWS ==========")
print(df.sample(5))
```

```python
# 11. Value Counts
print("\n========== VALUE COUNTS ==========")

# Automatically uses the first column
first_column = df.columns[0]
print(f"\nValue Counts for '{first_column}':")
print(df[first_column].value_counts())
```
