```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("ecommerce_sales_analytics_5000.csv")

# First 5 records
print("First Five Rows of Dataset:")
print(df.head())

# Shape
print("Shape of the Dataset:")
print(df.shape)

# Column names
print("Columns Names:")
print(df.columns)

# Data types
print("Datatypes:")
print(df.dtypes)

# Missing values
print("Missing Values:")
print(df.isnull().sum())

# Summary statistics
print("Summary Statistics:")
print(df.describe())

# Histogram
df.hist(figsize=(10,8))
plt.show()

# Bar Chart (First Categorical Column)
df["product_category"].value_counts().plot(kind="bar")
plt.title("Category Count")
plt.show()

# Pie Chart
df["product_category"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Category Distribution")
plt.ylabel("")
plt.show()

# Scatter Plot
plt.scatter(df["unit_price"], df["quantity"])
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.title("Price vs Quantity")
plt.show()

# Box Plot
plt.boxplot(df["revenue"])
plt.title("Box Plot of revenue")
plt.show()

# Heatmap
plt.imshow(df.corr(numeric_only=True), cmap="coolwarm",)
plt.colorbar()
plt.xticks(range(len(df.corr(numeric_only=True).columns)),
           df.corr(numeric_only=True).columns,
           rotation=90)
plt.yticks(range(len(df.corr(numeric_only=True).columns)),
           df.corr(numeric_only=True).columns)
plt.title("Correlation Heatmap")
plt.show()

# Summary
print("""
Summary:
1. Dataset loaded successfully.
2. Checked shape, columns, data types, and missing values.
3. Generated summary statistics.
4. Created Histogram, Bar Chart, Pie Chart, Scatter Plot, Box Plot, and Heatmap.
5. These graphs help understand the data distribution and relationships.
""")
```
